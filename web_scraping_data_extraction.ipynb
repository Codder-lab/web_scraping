# **Netflix Data**

**STEPS FOR EXTRACTING THE DATA**

1. Send an HTTP Request to the web page

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
```

```python
data = requests.get(url).text
#print(data)
```

2. Parse the HTML Content

```python
soup = BeautifulSoup(data, 'html5lib')
```

3. Identify the HTML Tags

```python
netflix_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])
```

4. Use a Beautiful Soup method for extracting data

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text

    # Create a temporary DataFrame for the new row
    temp_df = pd.DataFrame({"Date":[date], "Open":[Open], "High":[high], "Low":[low], "Close":[close], "Adj Close":[adj_close], "Volume":[volume]})

    # Use concat to append the new row to the main DataFrame
    netflix_data = pd.concat([netflix_data, temp_df], ignore_index=True)
```

Step 5: Print the Extracted Data

```python
netflix_data.head()
```

# **Amazon Data**

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html"
```

```python
html_data = requests.get(url).text
#print(html_data)
```

```python
soup = BeautifulSoup(html_data, "html5lib")
```

```python
soup.title
```

```python
amazon_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])

for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text

    temp_data = pd.DataFrame({"Date":[date], "Open":[Open], "High":[high], "Low":[low], "Close":[close], "Adj Close":[adj_close], "Volume":[volume]})
    amazon_data = pd.concat([amazon_data, temp_data], ignore_index=True)
```
